Global Learning System 2 : Numerical Integration for Lift Coefficient

# $$C_l = \int_{0}^1 (C_{p,upper} - C_{p,lower}) \ d(x/c) $$

```python
import numpy as np
import scipy.interpolate as interpolate

# Composite 1/3 Simpson Rule (Integration)
def simps (f, a, b, N=50):
  if N%2 != 0:
    return -1
  else:
    h = (b-a)/N
    x = np.linspace(a, b, N+1)
    y = f(x)
    S = h/3 * np.sum(y[0:-1:2] + 4*y[1::2] + y[2::2])
    return S

# Initialize
a = 0
b = 1
Cp_upper = interpolate.CubicSpline(xUpper[::-1], yUpper[::-1])
Cp_lower = interpolate.CubicSpline(xLower, yLower)

# Calculation
Cl_upper = simps(Cp_upper, a, b, 1000)
Cl_lower = simps(Cp_lower, a, b, 1000)


Cl = Cl_upper - Cl_lower
print(f'Cl = {Cl}')
```
